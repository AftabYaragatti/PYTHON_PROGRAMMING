```python
n1=int(input('Enter the First Number : '))
n2=int(input('Enter the Second Number : '))
# 1.Add two numbers.
add=n1+n2
print('Addition :',add)
# 2.Substract two numbers.
sub=n1-n2
print('Substraction : ',sub)
# 3.Multiplication two numbers.
mul=n1*n2
print('Multiplication :',mul)
# 4.Division two numbers.
div=n1/n2
print('Division :',div)
# 5.Modulas two numbers.
mod=n1%n2
print('Modulas :',mod)
# 6.Floor Division two numbers.
floor=n1//n2
print('Floor Division :',floor)
# 7.Power of two numbers.
power=n1**n2
print('Power :',power)

# 8.Concatenation of Two Strings.
s1=input('Enter String 1 :')
s2=input('Enter String 2 :')
concat=s1+s2
print('Concatenation :',concat)
# 9.Replication of Strings.
replica=s1*4
print('Replication :',replica)
```

```python
# 6.Convert total minutes into hours and remaining minutes.
minutes=int(input('Enter Total Minutes : '))
hrs=minutes//60 
print('Hours : ',hrs)
remMin=minutes%60
print('Remaining Minutes : ',remMin)
```

```python
# 8.Reverse a 3-digit number using % and //.
num=int(input('Enter the Number :'))
num1=num//100 # 123 - 123//100->1
num2=(num//10)%10 # (123//10)%10->(12)%10->2
num3=num%10 # 123%10->3
rev=num3*100+num2*10+num1
print(rev)
```

```python
# 9.Find sum of 3 digits of a number.
num=int(input('Enter the Number :'))
n1=num//100
n2=(num//10)%10
n3=num%10
print('Sum of Digits :',n1+n2+n3)
```

```python
# 10.Find product of digits of a number.
num=int(input('Enter the Number :'))
n1=num//100
n2=(num//10)%10
n3=num%10
print('Sum of Digits :',n1*n2*n3)
```

```python
# 11.Extract last digit of a number.
num=int(input('Enter the number :'))
LD=num%10
print("Last Digit :",LD)
```

```python
# 12.Remove last digit of a number.
num=int(input('Enter the number :'))
RLD=num//10
print("Last Digit :",RLD)
```

```python
# 13.Count1number of digits using //.
num=int(input('Enter the number :'))
count=0
while num:
    num=num//10
    count+=1
print("Digit Count :",count)
```

```python
# 14.Check whether a number is 3-digit or not using arithmetic logic.
num=int(input('Enter the number :'))
res=(num//100>0) and (num//1000==0)
print(f'Number {res} is 3 Digit Number')
```

```python
# Intermediate :
# 1.Swap two numbers using arithmetic operators (without third variable).
# 2.Convert total minutes into hours and remaining minutes.
# 3.Calculate compound interest and Simple interest.
# 4.Reverse a number using % and //.
# 5.Check Armstrong number (e.g., 153).
# 6.Check palindrome number.
# 7.Find GCD of two numbers using %.
# 8.Find LCM using arithmetic formula.
# 9.Calculate compound interest.
# 10.Calculate simple interest.
# 11.Calculate percentage of marks.
# 12.Create a simple calculator (+, -, *, /).
# 13.Convert days into years, months, and days.
# 14.Calculate average of n numbers.
```

```python
# 1.Swap two numbers using arithmetic operators (without third variable).
num1=int(input('Enter the First Number :'))
num2=int(input('Enter the Second Number :'))
print(f'Before Swap : {num1} and {num2}')
num1=num1+num2
num2=num1-num2
num1=num1-num2
print(f'After Swap : {num1} and {num2}')
```

```python
# 2.Convert total minutes into hours and remaining minutes.
Tminutes=int(input("Enter the Total Minutes :"))
hrs=Tminutes//60
print(hrs)
Rminutes=Tminutes%60
print(Rminutes)
```

```python
# 3.Calculate compound interest and Simple Interest.
Principal=int(input('Enter the Principal Value :'))
Rate=int(input('Enter the Rate :'))
Time=int(input('Enter the Time :'))
CI=Principal*((1+Rate)/100)-Principal
print(f'Compund Interest : {CI}')
SI=(Principal*Rate*Time)/100
print(f'Simple Interest : {SI}')
```

```python
# 4.Reverse a number using % and //.
num=int(input('Enter the Number :'))
num1=num//100 # 123 - 123//100->1
num2=(num//10)%10 # (123//10)%10->(12)%10->2
num3=num%10 # 123%10->3
rev=num3*100+num2*10+num1
print(rev)
```

```python
num=int(input("Enter 3 digit number: "))
rev=(num%10)*100+((num//10)%10)*10+(num//100)
print("Reversed:",rev)
```

```python
# 5.Check Armstrong number (e.g., 153).
num=int(input("Enter 3 digit number: "))
sum=(num%10)**3+((num//10)%10)**3+(num//100)**3
print("Armstrong Check:",sum)
```

```python
# 6.Check palindrome number.
num=int(input("Enter 3 digit number: "))
rev=(num%10)*100+((num//10)%10)*10+(num//100)
print("Palindrome Check:", rev)
```

```python
# 7.Find GCD of two numbers using %.
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
gcd=a if b==0 else b
print("GCD:", gcd)
```

```python
# 8.Find LCM using arithmetic formula.
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
lcm=(a*b)//1   # assuming gcd = 1
print("LCM:", lcm)
```

```python
# 9.Calculate compound interest.
p=float(input("Principal: "))
r=float(input("Rate: "))
t=float(input("Time: "))
ci=p*((1+r/100)**t)-p
print("Compound Interest:", ci)
```

```python
# 10.Calculate simple interest.
p=float(input("Principal: "))
r=float(input("Rate: "))
t=float(input("Time: "))
si=(p*r*t)/100
print("Simple Interest:", si)
```

```python
# 11.Calculate percentage of marks.
total=float(input("Total Marks: "))
obtained=float(input("Obtained Marks: "))
percentage=(obtained/total)*100
print("Percentage:", percentage)
```

```python
# 12.Create a simple calculator (+, -, *, /).
a=float(input("First number: "))
b=float(input("Second number: "))
result=a+b
print("Result:", result)
```

```python
# 13.Convert days into years, months, and days.
days=int(input("Enter days: "))
years=days//365
months=(days%365)//30
remaining=(days%365)%30
print(years,months,remaining)
```

```python
# 14.Calculate average of n numbers.
a=float(input())
b=float(input())
c=float(input())
avg=(a+b+c)/3
print("Average:", avg)
```
