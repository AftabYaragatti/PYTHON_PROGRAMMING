```python
# 1. Default Arguments :
# A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument.
```

```python
# Example for Default Arguements :
def Sum(num1,num2,num3=35): # By default num2=34 is assigned, even though we are not passed third arguement it will take by default
    result=num1+num2+num3
    print(result)
Sum(12,32)
```

```python
# 2. Keyword Arguments :
# In keyword arguments, values are passed by explicitly specifying the parameter names, so the order doesn’t matter.
```

```python
# Example for Keyword Arguemnet :
def Sum(num1,num2,num3): # No matter at what order we are passing variables
    result=num1+num2+num3
    print(f'Sum={result}')
Sum(num1=12,num2=23,num3=42)
Sum(num3=12,num1=23,num2=42)
```

```python
# 3. Positional Arguments :
# In positional arguments, values are assigned to parameters based on their order in the function call.
```

```python
# Example for Positional Arguemnet : 
def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Aftab", 21)

print("\nCase-2:")
nameAge(21, "Aftab") # If you change the order of parameter then it will assinged to those variable only
```

```python
# 4. Arbitrary Arguments :
# In Python Arbitrary Keyword Arguments, *args and **kwargs can pass a variable number of arguments to a function using special symbols. 
# There are two special symbols:
# *args in Python (Non-Keyword Arguments)
# **kwargs in Python (Keyword Arguments)
# Syntax : def function_name(*agrs,**kwargs):
#                  Block of code
#          function_name(inside this you can pass minimum 2 parameter, if you want ,you can pass more than 2 also)
```

```python
# Example for Arbitrary Arguemnet : 
def nameAgeMarks(*name,**marks): # here we arre not declare age parameter but we will get output, because of * we can do thi
    print(f'Name : {name}')
    print(f'Age : {age}')
    print(f'Marks : [{marks}]')

nameAgeMarks("Aftab",21, m1=87,m2=84,m3=85,m4=93,m5=87) # * means without declaring/passw=ing with keyword it will take this,
                                                        # ** means with keyword we nee to pass
```

```python
def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

# Function call with both types of arguments
myFun('Hey', 'Welcome', first='Aftab', mid='I', last='Yaragatti')
```

```python
# 5. Anonymous Functions [Return] :
# In Python, an anonymous function means that a function is without a name. 
# As we already know the def keyword is used to define the normal functions and the lambda keyword is used to create anonymous functions.
# Syntax : def function_name(parameter):
#                 return condition
```

```python
# Pass by Reference (mutable) and Pass by Value( immutable) :
# In Python, variables are references to objects. When we pass them to a function, 
# the behavior depends on whether the object is mutable (like lists, dictionaries) or immutable (like integers, strings, tuples).
# Mutable objects: Changes inside the function affect the original object.(# Pass by Reference)
# Immutable objects: The original value remains unchanged (# Pass by Value)
```

```python
# # Example of Pass by reference and Pass by value Function : Pass by Reference and Pass by Value

# Function modifies the first element of list : Pass By Value
def myFun(x):
    x[0] = 20
lst = [10, 11, 12, 13]
myFun(lst)   # Pass by Reference
print(lst)   # list is modified, 

# Function tries to modify an integer
def myFun2(x):
    x = 20
a = 10
myFun2(a)    # Pass by Value
print(a)     # integer is not modified
```

```python
# 6. Recursive Functions :
# A recursive function is a function that calls itself to solve a problem. 
# It is commonly used in mathematical and divide-and-conquer problems. 
# Always include a base case to avoid infinite recursion.
```

```python
# # Example of Recursive Function :
def factorial(n):
    if n == 0:  
        return 1
    else:
        return n * factorial(n - 1) 
print(factorial(4))
```

```python
# 7. Lambda Function :
# In Python, a lambda function is a small, anonymous function defined without using the def keyword.
# It can take any number of arguments but can only contain one expression, whose result is automatically returned.
# Syntax : FunctionName =  lambda arguement : expression/statement
```

```python
# # Example of Lambda Function :
res= lambda b : b*b
print(res(4))
```

```python
# 8. Closure Function :
# When inner function remembers the variables from its enclosing scope,even if the outer function has finished executing.
```

```python
# Example of Closure Function :
def Outer(x):
    def Inner(y):
        return x+y
    return Inner
Sum=Outer(20)
print(Sum(10))
```

```python
# 9. Decorator Function :
# That modifies another fuction behaviour withoud changing its code. Takes another function as a input parameter
```

```python
# Example of Decorator Function :
def decorator(func):
    def wrapper(*args,**kwargs): # we created wrapper() function inside decorator() , because we don't want directly to change the orginal func.
        print('Before Function Call')
        result=func(*args,**kwargs) # inside wrapper() original function calls.
        print('After Function Call')
        return result
    return wrapper # original function func=say_hello(), but decorator returns wrapper
                   #we are actually calling say_hello('Aftab') but wrapper('Aftab') get executes
@decorator  # Internally in Python it does : say_hello=decorator(say_hello)
def say_hello(name):
    print(f'Hello {name}')
say_hello(name='Aftab') # 'Aftab' - argument, name='Aftab' - Key word arguement 

# Steps :
# First @ decorator is called and passed say_hello() function as input parameter in decorator() dunction, i.e, decorator(say_hello)
# Inside Decotor we created an another function called wrapper() where it takes arguments
# Inside wrapper() we have called one fucntion called func() where it takes the argument from say_hello() function then prints the arguement
# new function wrapper() is returns to the decorator function
```

```python
def Outer(student):
    def wrapper(*args,**kwargs):
        print('Before Function Call')
        result=student(*args,**kwargs)
        print('After Function Call')
    return wrapper
@Outer 
def say_name(name):
    print(f'Name : {name}')
say_name('Aftab')

@Outer 
def say_age(age):
    print(f'Age : {age}')
say_age(21)
```

```python
## Use Cases of Lambda Function :
# 1. Using with Condition Checking
# 2. Using with List Comprehension
# 3. Using for Returning Multiple Results
# 4. Using with filter()
# 5. Using with map()
# 6. Using with reduce()
```

```python
# 1. Using with Condition Checking :
# A lambda function can use conditional expressions (if-else) to return different results based on a condition.
```

```python
check = lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Zero"
print(check(5))   
print(check(-3))  
print(check(0))
```

```python
# 2. Using with List Comprehension :
# Lambda functions can be combined with list comprehensions to apply the same operation to multiple values in a compact way.
```

```python
func = [lambda arg=x: arg * 10 for x in range(1, 5)]
for i in func:
    print(i())
```

```python
# 3. Using for Returning Multiple Results :
# Although a lambda can contain only one expression, it can still return multiple results by combining them into a tuple.
```

```python
calc = lambda x, y: (x + y, x * y, x - y)
res = calc(3, 4)
print(res)
```

```python
# 4. Using with filter() : Keeps only the elements that satisfys the condition
# filter() function uses a lambda expression to select elements from a list that satisfy a given condition, such as keeping only even numbers.
```

```python
c = [1, 2, 3, 4, 5, 6]
even = filter(lambda x: x % 2 == 0, c)
print(list(even))

# OR

c = [1,2,3,4,5,6]
even=list(filter(lambda x:x%2==0,c))
print(even)
```

```python
# 5. Using with map() : transform a list
# map() function applies a lambda expression to each element of a list and returns a new list with the transformed values.
```

```python
a = [1, 2, 3, 4]
square = map(lambda x: x * x, a)
print(list(square))
```

```python
# 6. Using with reduce()
# reduce() function repeatedly applies a lambda expression to elements of a list to combine them into a single result.
```

```python
from functools import reduce
a = [1, 2, 3, 4]
mul = reduce(lambda x, y: x * y, a)
print(mul)
```

```python
# lambda vs def Keyword : 
# In Python, both lambda and def can be used to define functions, but they serve slightly different purposes. 
# While def is used for creating standard reusable functions, lambda is mainly used for short, 
# anonymous functions that are needed only temporarily. 
```

```python
# Example :
# def Keyword Function
def Sum(x):
    return x+x
print(Sum(4))

# Lambda Function
Sum=lambda x : x+x
print(Sum(4))
```
