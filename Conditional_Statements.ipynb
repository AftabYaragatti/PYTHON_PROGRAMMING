```python
# Exampls :
# 1. Write the program to check whether you are pass with distinction,first class or second class or fail.
marks=int(input('Enter Your Marks :'))
if marks>=35:
    if marks>85:
       print(f'Distinction with {marks} marks')
    elif marks>75:
       print(f'First Class with {marks} marks')
    elif marks>55:
       print(f'Second Class with {marks} marks')
    elif marks>35:
       print(f'Pass with {marks} marks')
else:
    print(f'Fail with {marks} marks')
```

```python
# 14. Calculate electricity bill using slabs
# - 0-100 units: â‚¹1.50/unit
# - 101-200 units: â‚¹2.50/unit (after 100)
# - 201-300 units: â‚¹4/unit (after 200)
# - Above 300 units: â‚¹6/unit (after 300)
units=float(input('Enter units consumed :'))
if units<=100:
    bill=units*1.50
    print(f'Electricity Bill : {bill}')
elif units<=200:
    bill=100*1.50+(units-100)*2.50
    print(f'Electricity Bill : {bill}')
elif units <= 300:
    bill = 100 * 1.50 + 100 * 2.50 + (units - 200) * 4
    print(f'Electricity Bill : {bill}')
else:
    bill = 100 * 1.50 + 100 * 2.50 + 100 * 4 + (units - 300) * 6
    print(f'Electricity Bill : {bill}')
```

```python
# 21. Check whether a number is a leap year using nested if
year=int(input('Enter the Year :'))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f'{year} is an Leap Year')
        else:
            print(f'{year} is not a leap year')
    else:
            print(f'{year} is a leap year')
else:
    print(f'{year} is not a leap year')
```

```python
# 26. Find the quadrant of a coordinate (x, y)
# Logic : x>0 ,y>0 - x positive and y positive - 1st quadrant
# x<0 ,y>0 - x positive and y negative - 2nd quadrant
# x<0 , y<0 - x negative and y negative - 3rd quadrant
# x>0 ,y<0 - x positive and y negative - 4th quadrant
x=int(input('Enter X Coordinate Value :'))
y=int(input('Enter Y Coordinate Value :'))
if x>0 and y>0: # + +
    print(f'X and Y ({x,y}) in a First Quadrant')
elif x<0 and y>0: # - +
    print(f'X and Y ({x,y}) in a Second Quadrant')
elif x<0 and y<0: # - -
    print(f'X and Y ({x,y}) in a Third Quadrant')
elif x>0 and y<0: # + -
    print(f'X and Y ({x,y}) in a Fourth Quadrant')
else:
    print(f'X and Y ({x,y}) Invalid')
```
