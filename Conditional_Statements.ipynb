```python
# Conditional Statements : To make conditions in program we uses condtional statements ,based on True or False conditions.
# 1. If Condition : Execute block of code when condition is true.
# 2. If-Else Condition : Execute one block of code if condition is true,or another else block execute if it true.
# 3. If-Else-If Condition : used when multiple conditions want to check.
# 4. Nested If Statemeent : An id statement inside another if statement.
```

```python
# 1. If Conditional Statement :
# Syntax : if condition:
#              block of code
```

```python
# Examples :
# 1. Write a program to check given number is a positive number.
num=int(input('Enter the Number :'))
if num>=0:
    print(f'Number {num} is Positive Number')
```

```python
# 2. If-Else Conditional Statement :
# Syntax : if condition:
#              block of code
#          else:
#              block of code
```

```python
# Examples :
# 1. Write a program to check given number is a positive number or negative number.
num=int(input('Enter the Number :'))
if num>=0:
    print(f'Number {num} is Positive Number')
else:
    print(f'Number {num} is Negative Number')
```

```python
# 3. If-Else-If Conditional Statement :
# Syntax : if condition:
#              block of code
#          elif:
#              block of code
#          else:
#              block of code
```

```python
# Examples :
# 1. Write a program to check given number is a positive number or negative number or Zero.
num=int(input('Enter the Number :'))
if num>0:
    print(f'Number {num} is Positive Number')
elif num<0:
    print(f'Number {num} is Negative Number')
else:
    print(f'Number {num} is Zero')
```

```python
# 2. Write the program to check whether you are pass with distinction,first class or second class or fail.
marks=int(input('Enter Your Marks :'))
if marks>85:
    print(f'Distinction with {marks} marks')
elif marks>75:
    print(f'First Class with {marks} marks')
elif marks>55:
    print(f'Second Class with {marks} marks')
elif marks>35:
    print(f'Pass with {marks} marks')
else:
    print(f'Fail with {marks} marks')
```

```python
# 4. Nested-If Conditional Statement :
# Syntax : if condition:
#              if condition:
#                 block of code
#              elif condition:
#                 block of code
#              else:
#                 block of code
#          else:
#              block of code
```

```python
# Exampls :
# 1. Write the program to check whether you are pass with distinction,first class or second class or fail.
marks=int(input('Enter Your Marks :'))
if marks>=35:
    if marks>85:
       print(f'Distinction with {marks} marks')
    elif marks>75:
       print(f'First Class with {marks} marks')
    elif marks>55:
       print(f'Second Class with {marks} marks')
    elif marks>35:
       print(f'Pass with {marks} marks')
else:
    print(f'Fail with {marks} marks')
```

```python
# 4. Find the largest of three numbers
num1=int(input('Enter Fisrt Number :'))
num2=int(input('Enter Second Number : '))
num3=int(input('Enter Third Number : '))
if num1>num2: 
    if num1>num3:
       print(f'First Number {num1} is Greater Number among three numbers')
    else:
        print(f'Third Number {num3} is Greater Number among three numbers')
elif num2>num3:
    if num2>num1:
       print(f'Second Number {num2} is Greater Number among three numbers')
    else:
        print(f'Third Number {num3} is Greater Number among three numbers')
else:
    print(f'Third Number {num3} is Greater Number among three numbers')
```

```python
# 5. Check whether a year is a leap year
year=int(input('Enter Year :'))
if (year%400==0 and year%100==0):
    print(f'Year {year} is a Leap Year')
elif (year%4==0 and year%100!=0):
    print(f'Year {year} is a Leap Year')
else:
    print(f'Year {year} is a Not Leap Year')
```

```python
# 14. Calculate electricity bill using slabs
# - 0-100 units: â‚¹1.50/unit
# - 101-200 units: â‚¹2.50/unit (after 100)
# - 201-300 units: â‚¹4/unit (after 200)
# - Above 300 units: â‚¹6/unit (after 300)
units=float(input('Enter units consumed :'))
if units<=100:
    bill=units*1.50
    print(f'Electricity Bill : {bill}')
elif units<=200:
    bill=100*1.50+(units-100)*2.50
    print(f'Electricity Bill : {bill}')
elif units <= 300:
    bill = 100 * 1.50 + 100 * 2.50 + (units - 200) * 4
    print(f'Electricity Bill : {bill}')
else:
    bill = 100 * 1.50 + 100 * 2.50 + 100 * 4 + (units - 300) * 6
    print(f'Electricity Bill : {bill}')
```

```python
# 15. Check whether a triangle is valid : Sum of any 2 sides is equal to third side is valid traingle
side1=int(input('Enter First Side of Triangle :'))
side2=int(input('Enter Second Side of Triangle :'))
side3=int(input('Enter Third Side of Triangle :'))
if (side1+side2>side3 and side2+side3>side1 and side1+side3>side2):
    print(f'Triangle is Valid')
else:
    print(f'Triangle is InValid')
```

```python
# 16. Check whether a triangle is equilateral, isosceles, or scalene
# equilateral : a==b==c
# isosceles : a==b or b==c or a==c
# scalene : a,b,c
s1=int(input('Enter First Side of Triangle :'))
s2=int(input('Enter Second Side of Triangle :'))
s3=int(input('Enter Third Side of Triangle :'))
if s1==s2==s3:
    print(f'Triangle is Equilateral')
elif s1==s2 or s2==s3 or s3==s1:
    print(f'Triangle is Isosceles')
else:
    print(f'Triangle is Scalene')
```
