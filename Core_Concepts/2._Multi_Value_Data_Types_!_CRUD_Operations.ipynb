```python
## List  Data type with CRUD Operations :::
```

```python
# Basics of List : 
# List : It is collection of Homogeneous or Heterogeneous data/items/values/objects , enclosed between square brackets and separated by comma. 
# Syntax : variableName = [obj1,obj2,obj3,obj4]
# List is mutable datatype
# List supports indexing 
# We Can Perform Nested List Operation
```

```python
lst=[1,2,3,4,'Hello',23.9,'Python']
print(lst)
print(type(lst))
```

```python
#Adding two lists in single list
lst1=[1,2,'A','B']
lst2=[3,4,'C','D']
res=[lst1,lst2]
print(res)
```

```python
names=['Alice','Bob','Jhon','Robart']
print(names)
print(len(names))
```

```python
#To access the elements of list we uses indexing 
names=['Alice','Bob','Jhon','Robart']
print(names[2])
print(names[-1])
print(names[-1][2]) #to access the character inside the element
```

```python
## Set Data type with CRUD Operations :::
```

```python
# Basics of set : 
# Set : It is collection of Homogeneous or Heterogeneous data/items/values/objects , enclosed between curly braces which are immutable objects. 
# Syntax : variableName = {obj1,obj2,obj3,obj4} 
# Set itself is mutable but inside the set objects are immutable 
# Set does not supports indexing 
# Set is an Unordered collction 
# We Can't Pass Set inside another set because set itself is an mutable,only the objects are immuatable
```

```python
# Set Basics
s={1,2,3,'Hello'}
print(s,type(s))
```

```python
#we must and should only add immuatble objects inside the set
s={1,2,'Hello',(1,2),'P'} #tuple is also immutable hence it generates an hashcode for tuple data also, 
                          #where hashcode is used for immutable objects or dattypes such as tuple & string
print(s)
```

```python
#What if we add the mutable objects such as list and set inside the set object[set itself is mutable]
s={1,2,'Hello',[1,2],{2,3},'P'}
print(s) #we will get an error because list and set itself is an mutable objects,hence hashcode is not generated for list and set
```

```python
#Set Doesnot support Indexing
s=={1,2,'Hello',(1,2),'P'}
print(s)
print(s[2]) #will get an error because set doesnot support indexing
```

```python
# CRUD Operations of Sets ::

#Create Set :
# 1. Create/Define empty set objects :
s={}
print(s)
# 2. Create/Define set objects :
s1={4,1,2,3,(5,6),2,'Hello',5} #It doesnot follow order,it is an Unordered collection
print(s1)
```

```python
# Read the elements of Set :
# 1. Read entire set objects :
s={1,2,3,(5,6),'Hello'}
print(s)
# 2. Reas Specific set objects is not possible because set doesnot support indexing :
```

```python
# Update elements of set :
# As set is immuatble object it doesnot allow to be manipulate the objects, but we can update using built-in methods

# Delete elements of Set :
s={1,2,3,(5,6),'Hello'} #we can delete entire set items
print(s)
del s
print(s)
```

```python
## Frozenset Data type with CRUD Operations :::
```

```python
#Basics of Frozenset :
# Frozenset : It is collection of Homogeneous or Heterogeneous data/items/values/objects, which are immutable objects. 
# Syntax : variableName = frozenset(collection) #Collection may be any type but inside the collection objects must be immutable 
# Frozenset itself is immutable object
# Frozenset does not supports indexing 
# Frozenset is an Unordered Collection
# Frozenset stores only unique items , duplicate element will be removed
# We Can Pass Frozenset inside another frozenset because frozenset  is an immutable.
```

```python
# Basics of Frozenset
f=frozenset({1,2,'Hello',90,3,2}) #It doesnot follow order,it is an Unordered collection
print(f)
```

```python
#It does not support indexing 
f=frozenset({1,2,'Hello',90,3,2}) #It doesnot follow order,it is an Unordered collection
print(f)
print(f[2]) #get an error because it does not support indexing 
```

```python
# We Can Pass Frozenset inside another frozenset because frozenset  is an immutable.
f=frozenset({1,2,'Hello',90,3,2,frozenset({1,2})})
print(f)
```

```python
#If we pass mutable data types inside frozenset it gives error because only immuatable objects are allowed
f=frozenset({1,2,'Hello',90,[1,2,3],3,2,frozenset({1,2})})
print(f)
```

```python
#Collection may be any type but objects must be immuatable
f=frozenset([1,2,'Hello',90,3,2])
print(f)
```

```python
# CRUD Operations of Frozenset :

# Create and Read Frozenset :
f=frozenset({1,2,3,"hello"}) #create
print(f) #read
```

```python
# Update Frozenset is not possible becuase it is an immutable :

#Delete an Frozenset :
f=frozenset({1,2,3,"hello"})
print(f)
del f
print(f)
```

```python
## Dictionary Data type with CRUD Operations :::
```

```python
# Basics of Dictionary :
# Dictionary : It is collection of Homogeneous or Heterogeneous key-value pairs. 
# Syntax : variableName = {key1:value1,key2:value2.....keyn:valuen}
# Key-Value can be seprated by comma(,) and Key and Values are separated by colon (:)
# Key Should be Unique and It is an immutable data type
# Value can be any type 
# Dictionary always stores unique value, if key is repeated in dictionary then it will override the key of value
# It does not support indexing , isntead of it supports accessing values by using keys
```

```python
# Basic Dictionary :
d={1:"Python",2:"Java"}
print(d,type(d))
```

```python
# Accesing Dictionary Values
d={1:"Python",2:"Java"}
print(d)
print(d[1]) # access value by using key, If key doesnot exists it will through an error
print(d[1][3]) # Access specific value
```

```python
info={'Name':'Aftab','age':21,'Address':'Banglore','MobNo':6362099891,'subjects':['SQL','Python','Java'],1:(1,2,3),2:{1,2,3,4}}
print(info)
print(info[1]) 
print(info[2])
print(info['subjects'][1])
print(info['subjects'][1][2])
```

```python
# Create a new key value pair inside  a dictionary
info={'Name':'Aftab','age':21,'Address':'Banglore'}
print(info)
info['age']=23 # the key is already exists hence it will over rides with new value
print(info) 
```

```python
#Nested Dictionary :
info={'Name':'Aftab','age':21,'Address':'Banglore','marks':{'Python':230,'Java':342}}
print(info)
print(info['marks'])
print(info['marks']['Java'])
```

```python
# ReInitialization in Dictionary
d={1:100,2:200}
d[3]=400
print(d)
d=d[3]
print(d)
```

```python
# CRUD Operations of Dictionary:::

#Create Dictionary :
# 1. Create empty dictionary 
d={}
d1=dict({})
print(d,d1)
# 2. Create/define dictionary using dict() 
d2=dict({1:'Hello',2:300})
print(d2)
```

```python
#Read Dictionary :
# 1. Read entire dictionary 
d={1:'A',2:'B',3:'C'}
print(d)
# 2. Read only specific key values 
print(d[1])
```

```python
#Update Dictionary :
# 1. Update entire dictionary 
d={1:'A',2:'B',3:'C'}
print(d)
d1=dict({4:'D',5:'E',6:'F'})
print(d1)
# 2. Update specific key values 
d1[7]='G'
print(d1)
```

```python
#Delete Dictionary :
# 1. Delete entire dictionary 
d1={1:'A',2:'B',3:'C'}
print(d1)
del d1[3]
print(d1)
# 2. Update specific key values 
d2={1:'A',2:'B',3:'C'}
print(d2)
del d2
print(d2)
```

```python
## Range Data type :::
```

```python
# Basics of Range()
# It generates the series of numbers from start to end-1
# range is collection
# Syntax : varName=range(start=0:end:step=1)
# Immutable Objects
# Step Value is matters in range : If Step Value is Positive , Then condition must be start<end 
# If Step Value is Negative , Then condition must be start>end
# Note : No matters what is the start and end value, check always step value positive or negative, 
# If condition fails then return empty List/tuple/set/frozenset
```

```python
l=range(5) #start=0 end=5 #step=1 by default it will take
print(l) # I will get output in the form of range(0,5)
print(list(l)) # Displays the result interms of list
```

```python
l=range(1,6) #step=1 by default it will take
print(list(l))
```

```python
l=range(5,1)
print(list(l)) #Empty List
```

```python
l=range(1,9,1) #Step Value is positive, start<end Condition satisfy
print(list(l)) #we can convert at any time
print(tuple(l))
print(set(l))
print(frozenset(l))
```

```python
l=range(9,1,-1) #Step value is -1 means it decreasing the series by 1
print(list(l))
l1=range(20,10,-3)  #Step value is -3 means it decreasing the series by 3
print(list(l1))
l2=range(1,10,3)
print(list(l2))
```
