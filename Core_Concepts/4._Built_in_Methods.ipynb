```python
# String Built-in-methods :
# 1. string.upper() - String converts to uppercase
# 2. string.lower() - String converts to lowercase
# 3. string.capitalize() - String converts First letter as uppercase followed by lowercase
# 4. string.title() - The first letter of each word Converts uppercase in the string
# 5. string.swapcase() - That converts the upper case to lover case ,and vice-versa
# 6. string.strip() - Removes Blank/White spaces from strings
# 7. string.lstrip() - Removes Blank/White spaces from strings by left
# 8. string.rstrip() - Removes Blank/White spaces from strings by right
# 9. string.count(value) - Pass element as Parameter to count the number of element
# 10. string.find(value) - Pass element as Parameter to find the element
# 11. string.isalpha() - checks the string is contains alphabate or not
# 12. string.isdigit() - checks the string is contains digits or not
# 13. string.isupper() - checks the string is in uppercase or not
# 14. string.islower() - checks the string is in lowercase or not
# 15. string.isalnum() - checks the string is in containes alpha numeric or not
# 16. string.istitle() - checks the string is title or not
# 17. string.isspace() - checks the string is have spcae or not
# 18. string.startswith('strValue') - check whether strings starts with particular character or not
# 19. string.endswith('strValue') - check whether strings ends with particular character or not
# 20. string.replace('old str','new str') - String is replace with new string
# 21. joinedobj.join(string) - strings is joined by other string or char 
# 22. string.split(splited-str) - string is splitted by other char or str
# 23. len(str) - returns Length of the string
```

```python
s='Hello world! Welcome to python programming world'
print('Original String :',s)
print('Upper Case :',s.upper())
print('Lower Case :',s.lower())
print('Capitalize :',s.capitalize())
print('Title :',s.title())
print('Swapcase :',s.swapcase())
print('Strip :',s.strip())
print('Left Strip :',s.lstrip())
print('Rught Strip :',s.rstrip())
print('Find :',s.find('p'))
print('Count :',s.count('p'))
print('Is Alphabate :',s.isalpha())
print('Is Digit :',s.isdigit())
print('Is Alphabate Numeric :',s.isalnum())
print('Is Uppercase :',s.isupper())
print('Is Lowercase :',s.islower())
print('Is Space :',s.isspace())
print('Is Title :',s.istitle())
print('Starts with :',s.startswith('H'))
print('Starts with :',s.endswith('d'))
print('Replace :',s.replace('H','Y'))
print('Split :',s.split('o'))
print('Join :','o'.join(s))
print('Length :',len(s))
```

```python
## List Built-in-methods :
# 1. list.append() - Elements store at last index of list
# 2. list.pop() - Delete element from the last index of list
# 3. list.pop(index number) - Elements deleted from particular index
# 4. list.sort() - List sorts the elements by assending order
# 5. list.sort(reverse=True) - List sorts the elements by descending order
# 6. list.reverse() - List sorts the elements by descending order
# 7. list.insert(index,value) - It insert value from specific index with specific value
# 8. list.count(value) - It counts the particular count of value
# 9. list.remove(value/element) - It removes the particular element
# 10.list.clear() - List become empty
# 11.list.index(element) - return an index number for particular value
# 12.list.extend(another list) - one list extends an another list
# 13.del list - delete list using del keyword
# 14.list.copy() / list.copy(old list) - Shallow copy creates the new list but not new object 
# 15.list.deepcopy(old list) - Deep copy creates the new list and new object 
# 16.len(list) - returns Length of the list
```

```python
l=[12,1,34,3,6,1,8]
l.append(9)
print('Append :',l)
l.pop()
print('Pop :',l)
l.pop(3)
print('Pop with index :',l)
l.sort()
print('Sort :',l)
l.reverse()
print('Sort Reverse Order :',l)
l.sort(reverse=True)
print('Sort Reverse Order :',l)
l.insert(2,100)
print('Insert element with index : ',l)
a=l.count(1)
print('Counts :',a)
l.remove(1)
print('Remove particular element :',l)
b=l.index(100)
print('Index :',b)
l1=[1,2,3,4]
l2=[13,2,5,2]
l1.extend(l2)
print('Extended List :',l1)
l1.copy()
print('Shallow Copy :',l1)
c=len(l)
print('Length of List :',c)
l.clear()
print('Clear all elements :',l)
```

```python
# Set Built-in-methods :
# 1. set.add(element)
# 2. set.update(elements)
# 3. set.remove(element)
# 4. set.discard(element) - Discard if element is present
# 5. set.clear()
# 6. set.copy()
# 7. set.union() - takes no arguemnet
# 8. set.pop() - takes no arguement, removes first element
# 9. set1.issubset(set2)
# 10.set1.issuperset(set2)
# 11.set1.isdisjoint(set2)
# 12.Union =  set1 | set2
# 12.Intersection =  set1 & set2
# 12.Difference =  set1 - set2
# 12.Symmetric_Difference =  set1 ^ set2
```

```python
s={1,2,3,4,5,6}
s.add(8)
print('Add :',s)
s.update({9})
print('Update :',s)
s.remove(4)
print('Remove :',s)
s.discard(9)
print('Discard :',s)
s.copy()
print('Copy :',s)
s.pop()
print('Pop :',s)
s.union()
print('Union ;',s)
s.clear()
print('Clear :',s)
set1={1,2,3,4,9,7}
set2={2,5,7,3,1,8}
print(set1.issuperset(set2))
print(set1.isdisjoint(set2))
Union =  set1 | set2
print('Union :',Union)
Intersection =  set1 & set2
print('Intersection :',Intersection)
Difference =  set1 - set2
print('Difference :',Difference)
Symmetric_Difference =  set1 ^ set2
print('Symmetric_Difference :',Symmetric_Difference)
```

```python
# Frozenset Built-in-methods :
# 1. frozenset1.union(frozenset2) - returns unique element
# 2. frozenset1.intersection(frozenset2) - returns comman element
# 3. frozenset1.difference(frozenset2) - returns except comman elements from fset 1 and 2 , returns only fset1 element
# 4. frozenset1.symmetric_difference(frozenset2) - only returns unique element
# 5. frozenset1.issubset(frozenset2)
# 6. frozenset1.isdisjoint(frozenset2)
```

```python
f1=frozenset({1,2,3,4,5})
f2=frozenset({4,8,2,6,9})
print(f1.union(f2)) 
print(f1.intersection(f2))
print(f1.difference(f2))
print(f1.symmetric_difference(f2))
print(f1.issubset(f2))
print(f1.isdisjoint(f2))
```
