```python
# 1. For Loop :For loops is used to iterate over a sequence such as a list, tuple, string or range. 
# It allow to execute a block of code repeatedly, once for each item in the sequence.
# Syntax : for item in iteratable:
#              block of code
```

```python
# 1. For Loop Example :
s=input('Enter the String :')
for i in s:
    print(i)
```

```python
# 2. For Loop with Range Example :
num=int(input('Enter the Number :'))
for i in range(num):
    print(i)
```

```python
# 3. For Loop with Index, Value using Enumeracte Example  :
# Enumerate Syntax : for index,item in enumerate(iteratable):
#                            Block of code
s=input('Enter the String :')
for index,i in enumerate(s):
    print(index,i)
```

```python
# 2. While Loop : In Python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied. 
# When the condition becomes false, the line immediately after the loop in the program is executed.
# Syntax : while condition:
#             block of code
```

```python
# While Loop Example :
num=int(input('Enter the Number You want ro repeat :'))
count=0
while count<num:
    count+=1
    print('Hello World')
```

```python
# 3. Infinite while loop : If we want a block of code to execute infinite number of times then we can use the while loop in Python to do so.
```

```python
# 4. Nested Loop : Python programming language allows to use one loop inside another loop which is called nested loop. 
# Syntax : for item in iteratble :
#             for item in iteratable:
#                 block of code
```

```python
# Nested Loop Example :
for i in range(1, 5):
    for j in range(i):
        print(i, end=' ')
    print()
```

```python
# 5. Else Clause with For Loop :
# syntax : for item in iterable:
#             block of code
#          else:
#             block of code
```

```python
# Example 1 withoud break : 
for i in range(10):
    if i%2==0:
        print(f'{i} is Even Number')
    else:
        print(f'{i} is Odd Number')
else:
    print('Loop Completed withoud break')
```

```python
# Example 2 with break :
for i in range(1,10):
    if i%2==0:
        print(f'{i} is Even Number')
        break
else:
    print('Loop Completed withoud break')
```

```python
# 6. Else Clause with While Loop :
# syntax :  while condition:
#             block of code
#          else:
#             block of code
```

```python
# Example 1 withoud break :
i=0
while i<5:
    if i%2==0:
        print(f'{i} is Even Number')
    i+=1
else:
    print('Loop Completed withoud break')
```

```python
# Example 2 withoud break :
i=0
while i<5:
    if i%2==0:
        print(f'{i} is Even Number')
        break
else:
    print('Loop Completed withoud break')
```

```python
# Loop Control Statements :
# 1. Continue Statement
# 2. Break Statement
# 3. Pass Statement
```

```python
# 1. Continue Statement : The continue statement in Python returns the control to the beginning of the loop.
```

```python
# Example :
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# 2. Break Statement : The break statement in Python brings control out of the loop.
```

```python
# Example :
for i in range(10):
    if i==6:
        break
    print(i)
```

```python
# 3. Pass Statement : We use pass statement in Python to write empty loops. Pass is also used for empty control statements, functions and classes
```

```python
# Example :
for s in 'Hello World':
    pass
print('Last Letter :',s)
```

```python
# TERNARY : 
# Syntax : varName=IfStatementBlock if condition else ElseStatementBlock
```

```python
# Example :
n=int(input('Enter the Number'))
Odd='Odd Number' if n%2!=0 else 'Even Number'
print(Odd)
```
