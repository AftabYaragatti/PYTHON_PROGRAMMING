```python
### VARIABLES IN PYTHON ###
```

```python
# Syntax : variableName=data(any type)
# We can Initialization the varibles using different way as follows ::
# 1. Assign Single Value to a Single Variable
# 2. Re-Initialization of Value to a Variable
# 3. Assign Multiple Value to a Multiple Variable
# 4. Initializing multiple varible to a single data
# 5. Assign multiple values for single variable with the help of *  
# 6. Assigning One Variable to another varibale as a reference
# 7. Variable initialization Using Unpack Collection
```

```python
# 1. Assign Single Value to a Single Variable
a=100
print(a)
b='Hello World'
print(b)
```

```python
# 2. Re-Initialization of Value to a Variable
a=100
print(a)
a=200
print(a)
```

```python
# 3. Assign Multiple Value to a Multiple Variable
a,b,c=100,200,300
print(a,b,c)
```

```python
# 4. Initializing multiple varible to a single data
a=b=c=100
print(a,b,c)
```

```python
# 5. Assign multiple values for single variable with the help of *  
a,*b=100,200,300,400,500,600
print(a,b)
```

```python
# 6. Assigning One Variable to another varibale as a reference
a=100
b=a
print(a,b)
```

```python
# 7. Variable initialization Using Unpack Collection
names=['A','B','C']
x,y,z=names
print(x,y,z)
print(names)
```

```python
### DATA TYPES IN PYTHON ###
```

```python
## 1. Single Value Datatype ## 
# i)int 
# ii)float
# iii)complex
# iv)boolean
# v)none
```

```python
# i. int data type ::
a=100
print(a,type(a))
```

```python
# ii. float data type ::
b=13.30
print(b,type(b))
```

```python
# iii. complex data type ::
c=3+2j
print(c,type(c))
```

```python
# iv. boolean data type ::
d=True
print(d,type(d))
```

```python
# v. None Data type ::
e=None
print(e,type(e))
```

```python
## 2. Multi Value Datatype : 
# i.string 
# ii.list
# iii.tuple
# iv.range
# v.set
# vi.frozenset
# vii. dictionary
```

```python
# i. string data type :::
s='Hello'
print(s,type(s))
```

```python
# ii. list datatype :::
l=[10,20,30,40,50]
print(l,type(l))
```

```python
# iii. tuple datatype :::
t=(10,20,30,40,50)
print(t,type(t))
```

```python
# iv. set datatype :::
st={10,20,30,40,50,20}
print(st,type(st))
```

```python
# v. frozenset datatype :::
f=frozenset({10,20,30,40,50,20})
print(f,type(f))
```

```python
# vi. dictionary datatype :::
dict={'A':101,'B':102}
print(dict,type(dict))
```

```python
# vii. range datatype :::
r=range(1,20)
print(r,type(r))
```
