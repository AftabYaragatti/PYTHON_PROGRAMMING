```python
# 2. Relational Operator [Comparison Operators] : The operators compares values. It either returns True or False according to the condition.
```

```python
# Example of Comparison Operators :
a=10
b=20
# Equal to
print(a==b) #it return either true or false ,if both are same it return true else return as a false
# Not Equal to
print(a!=b) #it return either true or false ,if both are same it return true else return as a false

#another exapmle :
x=20
y='20'
print(x==y) # get an false , becuase x is an integre value, and y is an string 
print(x==int(y)) #get true, because y converts to int
print(str(x)==y) #get true, because x converts to string
```

```python
# Example of Relational Operators :
a=10
b=32
#less than
print(a<b) # True #it return either true or false,if condition satisfys.
#greater than
print(a>b) #false
#less than or equal to
print(a<=b) #True
#greater than or equal to
print(a>=b) #False
```

```python
# 3. Logical Operator : To perform Logical AND, Logical OR and Logical NOT operations we use logical operators in python. 
# It is used to combine conditional statements.
```

```python
# Example :

#Logical AND
a=True 
b=False
print(a and b)  # it combines two statement if both statements or data are True then ony it returns True , rather than this it return false

x=True #1
y=True #1
print(x and y) # If both statements are True, then it also true
print(a and b)  # it combines two statement either one statement should have True statements then ony it returns True , 
# rather than this it return false

#Logical OR
a1=True
b1=False
print(a1 or b1) #True, even though one statement should be thier as True

x1=False
y1=False
print(x1 or y1) # false, if both statements are False

#Logical NOT
a2=True
b2=False
print(not a2) #if statement is true , then not operator converts it to False and Vice-Versa
print(not b2) 
```

```python
# 4. Bitwise Operator : In Python Bitwise operators act on bits and perform bit-by-bit operations. These are used to operate on binary numbers.
```

```python
# Bitwise Operators in Python are as follows:
# Bitwise NOT  (~)
# Bitwise Shift (<<-left Shifr,>>-Right Shift)
# Bitwise AND (&)
# Bitwise XOR (^)
# Bitwise OR (|)
```

```python
#Example :
a = 5
b = 2
print(a & b) #a=101,b=010 => a+b = 000 => 0
print(a | b) #a=101,b=010 => a-b = 111 => 7
print(~b) # b=010,~101 =>-3
print(a ^ b) #q=101,b=010 => a^b => 111=>7 
#XOR stands for "exclusive or", and it returns true if and only if exactly one of the operands is true. 
# If the bits are different, it returns 1; otherwise, it returns 0.
#For Example 1-1=False,1-0=True,0-1=True,0-0=False
print(b ^ a)  #b=010,a=101 => a^b=111=>7
print(a >> 2) #Right SHift : # a=101, 010 101 => 010 010 =>2/2=1 
# If in case value or data is positve then take 0000 as initially in bits and for nagative value take 1111
# Bitwise Right Shift : Shifts the bits of the number to the right and fills 0 on voids left( fills 1 in the case of a negative number) as a result. 
# Similar effect as of dividing the number with some power of two.
print(a << 2) #Left Shift : # a=101, 010 101 => 010 010 =>2 ,2*2=4, 4*a=> 4*5=20
#Note In Bitwise Right Shift we need to divide the the decimal by given value.In Left Shift we need to do Multiply with given value

#Representation of numbers in bits
#decimalNumbers-Bits-Not Operator takes negative values
#0-000 = -8
#1-001 = -7
#2-010 = -6
#3-011 = -5
#4-100 = -4
#5-101 = -3
#6-110 = -2
#7-111 - -1
```

```python
a=2
print(a>>2) #a=010, bit of 2,bit of a = 010,010=>010,101 =>2/5=0.4 ,0.4/2=.2=0 #divide
print(a<<2) #a=010, => 010,010=>010,101 => 2*2=4,4*a=4*2=8
#010- 2 shift 2*2=4, then multiple by value of a - 4*2 = 8
```

```python
a=5
print(a>>3) #011,101=011,010=3,2=3/2=1.5,1.5/3=0.5=0
```

```python
# 6. Membership Operator :
# In Python, in and not in are the membership operators that are used to test whether a value or variable is in a sequence.
#in - True if value is found in the sequence
#not in - True if value is not found in the sequence
```

```python
#Example :
x=2
y=8
list=[1,2,3,4,5,6]
if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")
    
if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```

```python
# 7. Identity Operator :
#In Python, is and is not are the identity operators both are used to check if two values are located on the same part of the memory.
#Two variables that are equal do not imply that they are identical. 
# is - True if the operands are identical 
# is not - True if the operands are not identical 
```

```python
# Example : 
a=10
b=32
c=a
print(a is b) #False because a value is not present in b
print(a is c) #True because a value is present in b
```

```python
#Example on Tuple : Note : The type must be same to perform is and is not operator
t1=(1,2,3) 
t2=(4,5,2)
t3=t2
print(t1 is t2)
print(t2 is t3)
print(t1 is not t3)
```

```python
# 8. Ternary Operator :
#Ternary operators also known as conditional expressions are operators that evaluate something based on a condition being true or false
#Syntax : [on_true] if [expression] else [on_false] 
```

```python
# Example :
age=int(input('Enter Your Age :'))
"Eligible for vote" if age>=18 else "Not Eligible for Vote" 
```
