```python
# Slicing Operations in Python
```

```python
# Basics of Slicing ::
# Slicing :  It is the process of reading/writing multiple elements from collection
# Syntax : For Positive Indexing - varName[start=0:end=length(collection):step=1]
# For Negative Indexing - varName[start=-1:end=-length(collection):step=-1]
# It doesn't matter what value we have in start and end , we only see the step value,
# if it is positive then index start from left to right and if it is negative then index start from right to left 
# To get sequence of characters we using slicing
# In slicing we will not get any error, if will not find any index, then it will consider last index value in negative side 
```

```python
#Slicing Operations on List :
l=[1,2,3,4,5,6,'Hello','Python','Java']
print(l[:])
print(l[0:6])
print(l[6:0]) #empty
print(l[-1:-6]) # empty
print(l[-6:-1])
print(l[::1])
print(l[::-1])
print(l[1:7:2])
print(l[1:7:-2]) #empty
print(l[7:1:-2])
print(l[7][::-1])
```

```python
#Slicing Operations on Tuple :
t=(1,2,3,4,5,6,'Hello','Python','Java')
print(t[:])
print(t[0:6])
print(t[6:0]) #empty
print(t[-1:-6]) # empty
print(t[-6:-1])
print(t[::1])
print(t[::-1])
print(t[1:7:2])
print(t[1:7:-2]) #empty
print(t[7:1:-2])
print(t[7][::-1])
```

```python
#Slicing Operations on Dictionary :
d={1:'Hello',2:'Python',3:'Java'}
print(d)
print(d[1])
print(d[1][1:3])
print(d[2][::-1])
```
